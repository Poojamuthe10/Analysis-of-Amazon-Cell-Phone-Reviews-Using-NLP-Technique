```python
#importing required libraries..
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
cifar=tf.keras.datasets.cifar10
(training_images,training_labels),(testing_images,testing_labels)=cifar.load_data()
```

```python
names=['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
training_images.shape
```

```python
training_labels.shape
```

```python
training_images=training_images/255.0
testing_images=testing_images/255.0
```

```python
print(training_images, training_labels)
print(testing_images, testing_labels)
```

```python
plt.figure(figsize=(2,2))
plt.imshow(training_images[25000], cmap=plt.cm.binary)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.imshow(training_images[25000], cmap=plt.cm.binary)
plt.colorbar()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15,12))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(training_images[i], cmap=plt.cm.binary)
  plt.xlabel(names[training_labels[i][0]])
plt.show()
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(32,32,3)),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#training data to the model...
model.fit(training_images,training_labels,epochs=10)
```

```python
test_loss,test_acc=model.evaluate(testing_images,testing_labels)
```

```python
print("Test accuracy:",test_acc)
```
